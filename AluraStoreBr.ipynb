### Importa√ß√£o dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. An√°lise do faturamento


```python
#Criando um dicion√°rio (lojas_dict) onde cada chave √© o nome da loja e cada valor √© um DataFrame com os dados dessa loja.
#Isso permite percorrer todas as lojas usando for nome, df in lojas_dict.items(), simplificando o c√≥digo.

lojas_dict = {
    "Loja 1": loja,
    "Loja 2": loja2,
    "Loja 3": loja3,
    "Loja 4": loja4
}

#Calculando o faturamento total de cada loja (total da coluna Pre√ßo com .sum())
for nome, df in lojas_dict.items():
    faturamento = df["Pre√ßo"].sum()

    print(f"O faturamento total da loja {nome} √©: R$ {faturamento:,.2f}")

#Salvando esse total em um dicion√°rio para usar depois nos gr√°ficos
faturamentos = []
for nome, df in lojas_dict.items():
    faturamento = df["Pre√ßo"].sum()
    faturamentos.append(faturamento)


# Vari√°veis para apresenta√ß√£o anual
#Criei uma lista com os DataFrames das lojas.
#Converti a coluna de datas para o tipo datetime.
#Criei uma nova coluna com o ano da compra.
#Agrupei por ano e somei os pre√ßos.
#Salvei os totais em uma lista (faturamento_anual) com os resultados por loja.
#pd.to_datetime() para normalizar datas.
#.dt.year para extrair apenas o ano.
#.groupby() + .sum() para fazer agrega√ß√µes por categoria (ano).

dfs = [loja, loja2, loja3, loja4]
nomes = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
faturamento_anual = []

# Soma de cada loja anualmente
for df, nome in zip(dfs, nomes):
    df['Data da Compra'] = pd.to_datetime(df['Data da Compra'], dayfirst=True)
    df['Ano da Compra'] = df['Data da Compra'].dt.year
    faturamento = df.groupby('Ano da Compra')['Pre√ßo'].sum()
    faturamento_anual.append(faturamento)

    print(f"Faturamento anual da {nome}:")
    for ano, valor in faturamento.items():
        print(f"  Ano {ano}: R$ {valor:,.2f}")
    print()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Nomes das lojas e seus faturamentos
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [sum(f.values) for f in faturamento_anual]
valores_por_loja = [list(f.values) for f in faturamento_anual]
anos_todos = sorted(set().union(*[f.index for f in faturamento_anual]))

cores = ['#39A845', '#FF7F50', '#FFD700', '#FF69B4']

# Criar figura
fig, ax = plt.subplots(1, 2, figsize=(20, 5))

# Gr√°fico de Barras Horizontais ‚Äì Faturamento Total
bars = ax[0].barh(lojas, faturamentos, color=cores)
ax[0].set_title('Faturamento Total por Loja')
ax[0].xaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.2f}'))
ax[0].set_xlim(min(faturamentos) * 0.95, max(faturamentos) * 1.05)

# Linha da m√©dia
media_faturamento = sum(faturamentos) / len(faturamentos)
ax[0].axvline(media_faturamento, color='darkgreen', linestyle='--', linewidth=2,
              label=f'M√©dia: R$ {media_faturamento:,.2f}')
ax[0].legend()

# Anota√ß√µes nas barras
for bar in bars:
    width = bar.get_width()
    ax[0].text(width + 10000, bar.get_y() + bar.get_height() / 2,
               f'R$ {width:,.2f}', ha='left', va='center', fontsize=9)

# Gr√°fico de Linhas ‚Äì Faturamento Anual por Loja
for i, serie in enumerate(faturamento_anual):
    anos = serie.index
    valores = serie.values
    ax[1].plot(anos, valores, marker='o', label=lojas[i], color=cores[i])
    for ano, valor in zip(anos, valores):
        ax[1].text(ano, valor, f'{valor:,.0f}', color=cores[i], ha='center', va='bottom', fontsize=9)

ax[1].set_xticks(anos_todos)
ax[1].set_xticklabels(anos_todos)
ax[1].set_title('Faturamento Anual por Loja')
ax[1].legend(title='Lojas')
ax[1].yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.2f}'))

plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
# Lista com os DataFrames das lojas
lojas = [loja, loja2, loja3, loja4]

# Mostrar o Top 5 de categorias por loja
for idx, loja_df in enumerate(lojas, start=1):
    top5 = loja_df['Categoria do Produto'].value_counts().head(5)
    print(f'As categorias mais populares (Top 5) da Loja {idx} s√£o:\n{top5}')

    if idx < len(lojas):
        print()  # Linha em branco entre lojas
```

```python
import seaborn as sns

# Lista com os DataFrames das lojas
lojas = [loja, loja2, loja3, loja4]
venda_categorias = []

# Coletar os Top 5 de cada loja e armazenar como Series (index = categoria, values = contagem)
for loja_df in lojas:
    top5 = loja_df['Categoria do Produto'].value_counts().head(5)
    venda_categorias.append(top5)

# Combinar as Series em um DataFrame (categorias como √≠ndice, lojas como colunas)
df_top5 = pd.concat(venda_categorias, axis=1)
df_top5.columns = [f'Loja {i+1}' for i in range(len(lojas))]

# Substituir NaN por 0 (caso uma categoria n√£o esteja no Top 5 de todas as lojas)
df_top5.fillna(0, inplace=True)

# Criar o heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_top5, annot=True, cmap='Greens', fmt='g')
plt.title('Top 5 Categorias Mais Vendidas por Loja')
plt.ylabel('Categoria do Produto')
plt.xlabel('Loja')
plt.show()
```

# 3. M√©dia de Avalia√ß√£o das Lojas

```python
import pandas as pd

nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Criar um DataFrame vazio para juntar os dados
media_anual_df = pd.DataFrame()

for i, df in enumerate(dfs):
    medias_ano = df.groupby('Ano da Compra')['Avalia√ß√£o da compra'].mean().round(2)
    media_anual_df[nomes_lojas[i]] = medias_ano

print("M√©dia Anual da Avalia√ß√£o por Loja:\n")
print(media_anual_df.to_string())

# Criar DataFrame para as quantidades de avalia√ß√µes
avaliacao_notas_df = pd.DataFrame()

for i, df in enumerate(dfs):
    counts = df['Avalia√ß√£o da compra'].value_counts().sort_index()
    avaliacao_notas_df[nomes_lojas[i]] = counts

avaliacao_notas_df = avaliacao_notas_df.fillna(0).astype(int)

print("\nQuantidade de Avalia√ß√µes por Nota:\n")
print(avaliacao_notas_df.to_string())
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Supondo que dfs √© a lista dos seus dataframes [loja, loja2, loja3, loja4]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Pegando todas as notas poss√≠veis (de 1 a 5, por exemplo)
notas_possiveis = [1, 2, 3, 4, 5]

# Criar DataFrame vazio para armazenar contagem das avalia√ß√µes
avaliacao_notas_df = pd.DataFrame(index=notas_possiveis)

# Preenchendo o DataFrame com contagem de avalia√ß√µes para cada loja
for i, df in enumerate(dfs):
    contagem = df['Avalia√ß√£o da compra'].value_counts()
    # Garantir que todas as notas poss√≠veis estejam presentes no √≠ndice
    contagem = contagem.reindex(notas_possiveis, fill_value=0)
    avaliacao_notas_df[nomes_lojas[i]] = contagem

n_lojas = len(nomes_lojas)
n_notas = len(notas_possiveis)


cores = ['#39A845', '#FF7F50', '#FFD700', '#FF69B4']
x = np.arange(n_notas)  # posi√ß√£o no eixo x para cada nota
largura = 0.2  # largura das barras

fig, ax = plt.subplots(figsize=(10, 6))

# Desenhar barras agrupadas para cada loja
for i in range(n_lojas):
    ax.bar(x + i*largura, avaliacao_notas_df[nomes_lojas[i]], width=largura, label=nomes_lojas[i], color=cores[i])

# Configura√ß√µes do gr√°fico
ax.set_xticks(x + largura * (n_lojas - 1) / 2)
ax.set_xticklabels(notas_possiveis)
ax.set_xlabel('Nota da Avalia√ß√£o')
ax.set_ylabel('Quantidade de Avalia√ß√µes')
ax.set_title('Distribui√ß√£o das Avalia√ß√µes por Loja')
ax.legend(title='Lojas')

# Mostrar valor no topo das barras para facilitar leitura
for i in range(n_lojas):
    for j in range(n_notas):
        valor = avaliacao_notas_df[nomes_lojas[i]].iloc[j]
        ax.text(x[j] + i*largura, valor + 5, str(valor), ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
#Quantidade de produtos vendidos
venda_produtos = [
    loja['Produto'].value_counts(),
    loja2['Produto'].value_counts(),
    loja3['Produto'].value_counts(),
    loja4['Produto'].value_counts()]

#Variav√©is para lista de produtos mais e menos vendidos
produtos_mais_vendidos = []
produtos_menos_vendidos = []

#Extra√ß√£o dos produtos mais e menos vendidos por loja
for vendas in venda_produtos:
    max_vendas = vendas.max()
    min_vendas = vendas.min()
    produtos_max = vendas[vendas == max_vendas]
    produtos_min = vendas[vendas == min_vendas]
    produtos_mais_vendidos.append(produtos_max)
    produtos_menos_vendidos.append(produtos_min)

#Apresenta√ß√£o do resultado
for i in range(len(produtos_mais_vendidos)):
    print(f"Loja {i + 1} - Produtos mais vendidos:")
    for produto, quantidade in produtos_mais_vendidos[i].items():
        print(f" - {produto}: {quantidade} vendas")
    print()

    print(f"Loja {i + 1} - Produtos menos vendidos:")
    for produto, quantidade in produtos_menos_vendidos[i].items():
        print(f" - {produto}: {quantidade} vendas")
    print()
    if i > 2:
        continue
    print("-" * 40)
    print()
```

```python
import pandas as pd
import plotly.express as px

# Lista com nomes das lojas
lojas_nomeadas = [f'Loja {i+1}' for i in range(len(venda_produtos))]

# Montar os dados dos produtos mais e menos vendidos
dados = []

for i in range(len(produtos_mais_vendidos)):
    for produto, qtd in produtos_mais_vendidos[i].items():
        dados.append({'Loja': lojas_nomeadas[i], 'Produto': produto, 'Tipo': 'Mais Vendido', 'Quantidade': qtd})
    for produto, qtd in produtos_menos_vendidos[i].items():
        dados.append({'Loja': lojas_nomeadas[i], 'Produto': produto, 'Tipo': 'Menos Vendido', 'Quantidade': qtd})

# Criar DataFrame
df_vendas_extremos = pd.DataFrame(dados)

# Treemap com cores v√°lidas para o Plotly
fig = px.treemap(
    df_vendas_extremos,
    path=['Loja', 'Produto'],
    values='Quantidade',
    color='Quantidade',
    color_continuous_scale='rainbow',
    title='Produtos Mais e Menos Vendidos por Loja'
)

fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))
fig.show()
```
